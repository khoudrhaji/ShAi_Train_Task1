```python
import pandas as pd
import numpy as np

# Load my dataset


sal_data  = pd.read_csv("SalariesX.csv")
sal_data .head()
```

```python
sal_data.columns
```

# 1. Basic Data Exploration

```python
# part 1: Identify the number of rows and columns
num_rows, num_columns = sal_data.shape
print(f"Num of rows: {num_rows}")
print(f"Num of columns: {num_columns}")

# part 2: Determine the data types of each column
data_types = sal_data.dtypes
print("\nData types of each column:")
print(data_types)

# part 3: Check for missing values in each column
missing_values = sal_data.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

# 2. Descriptive Statistics

```python
list_of_name = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']

for column_name in list_of_name:
    mean_sal = sal_data[column_name].mean()
    median_sal  = sal_data[column_name].median()
    mode_sal  = sal_data[column_name].mode()[0]
    min_sal  = sal_data[column_name].min()
    max_sal  = sal_data[column_name].max()
    range_sal  = max_sal  - min_sal 
    std_dev_sal  = sal_data[column_name].std()
    
    print("#############################################")
    print(f"{column_name}")
    print(f"Mean Salary: {mean_sal}")
    print(f"Median Salary: {median_sal}")
    print(f"Mode Salary: {mode_sal}")
    print(f"Minimum Salary: {min_sal}")
    print(f"Maximum Salary: {max_sal}")
    print(f"Range of Salaries: {range_sal}")
    print(f"Standard Deviation of Salary: {std_dev_sal}")
    print("#############################################")
```

# 3. Data Cleaning

```python
sal_data_cleaned = sal_data.drop(columns=['Notes'])

for column in ['BasePay', 'OvertimePay', 'OtherPay']:
    median_value = sal_data_cleaned[column].median()
    sal_data_cleaned[column].fillna(median_value, inplace=True)
    
sal_data_cleaned=sal_data_cleaned.drop_duplicates()

print(sal_data_cleaned.isnull().sum())

cleaned_file_path = 'Sal_Cleaned.csv'
sal_data_cleaned.to_csv(cleaned_file_path, index=False)
```

# 4.Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Histogram for Salary Distribution
plt.figure(figsize=(10, 6))
sns.histplot(sal_data_cleaned['BasePay'], bins=30, kde=False)
plt.title('Distribute of Salaries')
plt.xlabel('Salaries')
plt.ylabel('Num of Employees')
plt.show()
```

```python
department_num = sal_data_cleaned['JobTitle'].value_counts().head(20)
plt.figure(figsize=(25, 25))
plt.pie(department_num, labels=department_num.index, autopct='%1.5f%%', startangle=160)
plt.title('Percentage % of Employees in JobTitle')
plt.show()
```

# 5. Grouped Analysis

```python
import pandas as pd

grp_col = ['JobTitle', 'Year']
grp_data = sal_data_cleaned.groupby(grp_col)
summary_status = grp_data['BasePay'].describe()
mean_sal = summary_status['mean'].sort_values(ascending=False)

print("Summary Statistics for each Group:\n")
print(summary_status)
print("\n Avg Salaries Compare with :\n")
print(mean_sal)
```

# 6. Simple Correlation Analysis

```python
comparison_column = 'OvertimePay'
correlation = sal_data['BasePay'].corr(sal_data[comparison_column])
plt.figure(figsize=(10, 6))
plt.scatter(sal_data['BasePay'], sal_data[comparison_column], alpha=0.5)
plt.title(f'Scatter Plot of BasePay vs. {comparison_column}')
plt.xlabel('BasePay')
plt.ylabel(comparison_column)
plt.show()
print(f"Correlation coefficient between BasePay and {comparison_column}: {correlation}")
```
